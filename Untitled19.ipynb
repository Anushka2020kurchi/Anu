```python
import seaborn as sns
from matplotlib import pyplot as plt
import warnings
%matplotlib inline
sns.set(style="white", context="talk")
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.filterwarnings("ignore",category=plt.cbook.mplDeprecation)
```

How many brands are available in the data?

```python
x=data['Brand'].value_counts()
print(x)
plt.figure(figsize=(8, 6))
fig=sns.barplot(x.index,x.values)
plt.xlabel('Brands')
plt.ylabel('Density')
fig.set_title('Distribution of Brands')
```

Which country does the &#39;Coffee House Holdings&#39; brand use?

```python
HouseC=data[data['Brand']=='Coffee House Holdings']['Country'].value_counts()
print(HouseC)
```

Which country does the &#39;Tevana&#39; brand use?

```python
TevanaC=data[data['Brand']=='Teavana']['Country'].value_counts()[:10]
print(TevanaC)
plt.figure(figsize=(8, 6))
fig=sns.barplot(TevanaC.index,TevanaC.values)
plt.xlabel('Country')
plt.ylabel('Density')
fig.set_title('Distribution of Tevana Brand')
```

Which country does &#39;Starbucks&#39; brand use?

```python
Starbucks=data[data['Brand']=='Starbucks']['Country'].value_counts()[1:10]
print(Starbucks)
plt.figure(figsize=(8, 6))
fig=sns.barplot(Starbucks.index,Starbucks.values)
plt.xlabel('Country')
plt.ylabel('Density')
fig.set_title('Distribution of Starbucks Brand')
```

Which country has Joint Venture ownership?

```python
JointVenture=data[data['Ownership Type']=='Joint Venture']['Country'].value_counts()
print(JointVenture)
plt.figure(figsize=(8, 6))
fig=sns.barplot(JointVenture.index,JointVenture.values)
plt.xlabel('Country')
plt.ylabel('Density')
fig.set_title('Distribution of JointVenture')
```

Which country has Licensed ownership?

```python
Licensed=data[data['Ownership Type']=='Licensed']['Country'].value_counts()
print('Total number of country is ', len(Licensed))
#Top ten countries
plt.figure(figsize=(8, 6))
fig=sns.barplot(Licensed.index[1:10],Licensed.values[1:10])
plt.xlabel('Country')
plt.ylabel('Density')
fig.set_title('Distribution of Licensed')
```

Which country has Company Owned ownership?

```python
CompanyOwned=data[data['Ownership Type']=='Company Owned']['Country'].value_counts()
print(CompanyOwned)
print('Total number of country is ', len(CompanyOwned))
#Top ten countries
plt.figure(figsize=(8, 6))
fig=sns.barplot(CompanyOwned.index[1:10],CompanyOwned.values[1:10])
plt.xlabel('Country')
plt.ylabel('Density')
```

How many cities are Starbucks stores available?

```python
City=data['City'].value_counts()
print(City)
print('Total number of cities is',len(City))
plt.figure(figsize=(12,8))
fig=sns.barplot(City.index[:10],City.values[:10])
plt.xlabel('City')
plt.ylabel('Density')
fig.set_title('Distribution of City')
```

How many states or province are Starbucks stores available?

```python
State=data['State/Province'].value_counts()
print(State)
print('Total number of states is',len(State))
plt.figure(figsize=(12,8))
fig=sns.barplot(State.index[:10],State.values[:10])
plt.xlabel('State')
plt.ylabel('Density')
fig.set_title('Distribution of State')
```
